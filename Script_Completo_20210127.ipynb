```python
#IMPUTACION: calculo de las modas para cada id
variables_imp = ['id_cli','estado_civil','nivel_academico','profesion','ocupacion','ciudad_residencia','departamento_residencia','ciudad_laboral','departamento_laboral']
```

```python
#Validacion de id_cli repetidos con valores para imputar
dict_id = datos['id_cli'].value_counts().to_dict()
dict_id
```

```python
#Encontrar Ids de prueba
id_imputacion = []
for key,value in dict_id.items():
    if ((datos.loc[datos['id_cli']== key,'profesion'].isnull()).sum()) != 0:
        if ((datos.loc[datos['id_cli']== key,'profesion'].isnull()).sum()) < value:
            id_imputacion.append(key)

id_imputacion
```

```python
#Encontrar Ids de prueba vacios todos
for key,value in dict_id.items():
    if ((datos.loc[datos['id_cli']== key,'profesion'].isnull()).sum()) == value:
        print(key)
        break
```

```python
#Id de prueba
datos[datos['id_cli'] == id_imputacion[0]]
```

```python
#Creacion de las modas de las variables por id
df_modas = datos[variables_imp]
df_modas_imp = df_modas.groupby(['id_cli']).agg(lambda x:x.value_counts().index[0])
```

```python
#remplazar los valores faltantes desde otros registros del id_cli
for index, row in datos.iterrows():
    if row.loc['id_cli'] in id_imputacion:
        for index_2, value in row.iteritems():
            if (pd.notna(value) == False) and (index_2 in variables_imp):
                row.loc[index_2] = df_modas_imp.loc[row.loc['id_cli']][index_2]
            else:
                pass
        datos.loc[index] = row
```

```python
scaler=StandardScaler()
df = datos.drop(['gasto_familiar'], axis=1) # quito la variable dependiente "Y"
scaler.fit(df) # calculo la media para poder hacer la transformacion
X_scaled=scaler.transform(df)# Ahora si, escalo los datos y los normalizo
```

```python
#Instanciamos objeto PCA y aplicamos
pca=PCA(.90)#otra opción es instanciar pca sólo con dimensiones nuevas hasta obtener un mínimo "explicado" ej.: pca=PCA(.85)
pca.fit(X_scaled) # obtener los componentes principales
X_pca=pca.transform(X_scaled)# convertimos nuestros datos con las nuevas dimensiones de PCA
principaldf = pd.DataFrame(data = X_pca)
finaldf = pd.concat([principaldf, datos[['gasto_familiar']]],axis = 1)
```

```python
print("shape of X_pca", X_pca.shape)
expl = pca.explained_variance_ratio_
print(expl)
print('suma:',sum(expl[0:100]))
#Vemos que con 5 componentes tenemos algo mas del 85% de varianza explicada
```

```python
#graficamos el acumulado de varianza explicada en las nuevas dimensiones
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
plt.show()
```
