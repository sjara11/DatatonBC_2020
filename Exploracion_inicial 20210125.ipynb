### Tratamiento de valores faltantes

```python
#Variables a imputar con valores del id en otro periodo
variables_imp = ['id_cli','estado_civil','nivel_academico','profesion','ocupacion','categoria','ciudad_residencia','departamento_residencia','ciudad_laboral','departamento_laboral'] 
```

```python
#Validacion de id_cli repetidos con valores para imputar
dict_id = datos['id_cli'].value_counts().to_dict()
dict_id
```

```python
#Encontrar Ids de prueba
id_imputacion = []
for key,value in dict_id.items():
    if ((datos.loc[datos['id_cli']== key,'profesion'].isnull()).sum()) != 0:
        if ((datos.loc[datos['id_cli']== key,'profesion'].isnull()).sum()) < value:
            id_imputacion.append(key)

id_imputacion
            
    
```

```python
#Encontrar Ids de prueba vacios todos
for key,value in dict_id.items():
    if ((datos.loc[datos['id_cli']== key,'profesion'].isnull()).sum()) == value:
        print(key)
        break
```

```python
#Id de prueba
datos[datos['id_cli'] == id_imputacion[0]]
```

```python
#Creacion de las modas de las variables por id
df_modas = datos[variables_imp]
df_modas_imp = df_modas.groupby(['id_cli']).agg(lambda x:x.value_counts().index[0])
```

```python
#remplazar los valores faltantes desde otros registros del id_cli
for index, row in datos.iterrows():
    if row.loc['id_cli'] in id_imputacion:
        for index_2, value in row.iteritems():
            if (pd.notna(value) == False) and (index_2 in variables_imp):
                row.loc[index_2] = df_modas_imp.loc[row.loc['id_cli']][index_2]
            else:
                pass
        datos.loc[index] = row
```

```python
#Validar la imputacion
datos[datos['id_cli']==id_imputacion[0]]
```

## Analisis estadistico de las variables

```python
#Analisis descriptivo de los datos
#Variables numericas
 
correlation_mat = datos.corr()
 
# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(correlation_mat, dtype=bool))
 
# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))
 
# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)
 
# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(correlation_mat, mask=mask,vmax = 1, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
plt.show()
```
